```python
def even_or_odd (n):
    if n%2 == 0 and n in range(2,6):
        print ('Not Weird')
    elif n%2 == 0 and n in range(6,21):
        print ('Weird')
    elif n%2 == 0 and n > 20:
        print ('Not Weird')
    else:
        print('Weird')

even_or_odd(1)
even_or_odd(2)
even_or_odd(6)
even_or_odd(20)
even_or_odd(21)
```
